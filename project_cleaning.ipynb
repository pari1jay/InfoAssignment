```python
import pandas as pd

#total number of sales
total_sales = data_2['sales'].sum()

#total volume of sales
total_volume = data_2['volume'].sum()

#number of unique cities
num_cities = data_2['city'].nunique()

print("Total number of sales:", total_sales)
print("Total volume of sales:", total_volume)
print("Number of unique cities:", num_cities)
```

```python
txh = data_2
# mean_price-mean sale price of all sales in a city in a given month
txh['mean_price'] = txh['volume'] / txh['sales']

# median sale price in a given month will generally be different from the mean_price- diff is given by price_diff
txh['price_dif'] = txh['mean_price'] - txh['median']

# calculates the proportion of listings that resulted in sales in a given month.
txh['sales_prop'] = txh['sales'] / txh['listings']
txh.head()
```

```python
sales_prop_greater_one = txh[txh['sales_prop'] > 1]
sales_prop_greater_one
```

```python
# median of 'price_dif' for each city
median_pricediff = data_2.groupby('city')['price_dif'].median().reset_index()

# Sorting in descending order 'med_price_dif'
median_pricediff = median_pricediff.sort_values(by='price_dif', ascending=False)

median_pricediff
```

```python
grouped_data = txh.groupby(['city', 'year'])
mean_monthly_sales = grouped_data['sales'].mean()
median_median_price = grouped_data['median'].median()
median_mean_price = grouped_data['mean_price'].median()
median_price_dif = grouped_data['price_dif'].median()

result_df = pd.DataFrame({
    'mean_monthly_sales': mean_monthly_sales,
    'median_median_price': median_median_price,
    'median_mean_price': median_mean_price,
    'median_price_dif': median_price_dif
}).reset_index()
result_df.head()
```

```python
import matplotlib.pyplot as plt

grouped_data = result_df.groupby(['year', 'city'])['mean_monthly_sales'].mean().unstack()
plt.figure(figsize=(12, 12))
#line plot with a different color line for each city
grouped_data.plot(marker='o')

plt.xlabel('Year')
plt.ylabel('Average Monthly Sales')
plt.title('Average Monthly Sales per Year for Each City')

plt.legend(title='City', labels=grouped_data.columns, loc='upper right')
plt.show()
```
